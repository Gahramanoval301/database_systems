## Section 1: Introduction to Data Mining & Decision Trees
Data mining is the process of semi-automatically analyzing large databases to find useful patterns. It is part of the larger area of Knowledge Discovery in Databases (KDD).

One fundamental predictive mechanism is **Classification**, where items are assigned to discrete classes. A common way to build classification models is using **Decision Trees**. To build a good tree, we need to find the "Best Split" for the data.

### The Gini Measure of Purity
The purity of a set of training instances can be measured using the Gini index. 
Formula: `Gini(S) = 1 - sum(p_i^2)` where `p_i` is the fraction of instances in class `i`.

### Exercise 1: Calculate Gini Impurity
Assume we have a node in a decision tree with 100 items. 70 belong to Class A (p_A = 0.70) and 30 belong to Class B (p_B = 0.30). Calculate the Gini impurity.

```python
def calculate_gini(p_a, p_b):
    gini_impurity = 1 - (p_a**2 + p_b**2)
    return gini_impurity

p_class_a = 0.70
p_class_b = 0.30

node_gini = calculate_gini(p_class_a, p_class_b)
print(f"Gini Impurity for the node: {node_gini:.2f}") # I did .2f for making it 0.42 instead of 0.42000..004
# Expected output: 0.42
```

## Section 2: Entropy and Information Gain
Another measure of purity is **Entropy**.
Formula: `Entropy(S) = -sum(p_i * log2(p_i))`

When a dataset `S` is split into subsets, we evaluate the split using **Information Gain**. We want to maximize the information gain, which means we want the resulting subsets to be as pure as possible.

### Exercise 2: Calculate Entropy
Use the `math.log2` function to calculate the entropy of the same node from Exercise 1.

```python
import math

def calculate_entropy(p_a, p_b):
    # Handle the edge case where a probability is 0 (log2(0) is undefined)
    if p_a == 0 or p_b == 0:
        return 0.0
    
    # Task completion
        
    entropy_val = 0.0
    
    if p_a > 0:
        entropy_val -= p_a * math.log2(p_a)
    if p_b > 0:
        entropy_val -= p_b * math.log2(p_b)
        
    return entropy_val

node_entropy = calculate_entropy(p_class_a, p_class_b)
print(f"Entropy for the node: {node_entropy:.4f}")
# Expected output: ~0.8813
```

## Section 3: Association Rules (Market Basket Analysis)
Retail shops look for associations between items. An association rule looks like: `Antecedent => Consequent` (e.g., `Bread => Milk`).

Two key metrics govern these rules:
1. **Support**: The fraction of the total population that satisfies BOTH the antecedent and consequent. 
2. **Confidence**: How often the consequent is true WHEN the antecedent is true.

### Exercise 3: Calculating Support and Confidence
You are given a list of 5 store transactions. We want to evaluate the rule: `Apples => Bananas`.

```python
# List of transactions (each transaction is a set of items bought)
transactions = [
    {'Apples', 'Bananas', 'Milk'},
    {'Apples', 'Milk'},
    {'Apples', 'Bananas', 'Bread'},
    {'Bananas', 'Bread'},
    {'Apples', 'Bananas'}
]

total_transactions = len(transactions)

# Task Completion

count_apples_and_bananas = 0
for t in transactions:
    if {'Apples', 'Bananas'}.issubset(t):
        count_apples_and_bananas += 1

count_apples = 0
for t in transactions:
    if 'Apples' in t:
        count_apples += 1

support = count_apples_and_bananas / total_transactions
confidence = count_apples_and_bananas / count_apples


print(f"Support for 'Apples => Bananas': {support}")
print(f"Confidence for 'Apples => Bananas': {confidence}")
```

## Section 4: The A Priori Technique
To find association rules efficiently, we use the **A Priori technique**. Generating all possible item combinations takes too much memory. 

**Optimization Principle**: Once an itemset is eliminated because its support is too small, NONE of its supersets need to be considered. If people rarely buy `Milk`, they definitely rarely buy `Milk + Screwdrivers`.

### Exercise 4: Pruning Candidate Itemsets
Filter a dictionary of itemset counts based on a minimum support threshold.

```python
itemset_counts = {
    'Milk': 450,
    'Bread': 500,
    'Screwdriver': 5,
    'Eggs': 300,
    'Hammer': 12
}

total_sales = 1000
min_support_threshold = 0.05 # 5% minimum support

large_itemsets = {}

# Task completion
for item, count in itemset_counts.items():
    support = count / total_sales
    if support >= min_support_threshold:
        large_itemsets[item] = support


print("Large Itemsets (Surviving Candidates for the next pass):")
print(large_itemsets)
```

## Section 5: Clustering
Clustering involves finding groups of points in data such that similar points lie in the same cluster. 

A standard approach is grouping points into `k` sets such that the average distance of points from the **centroid** (the center point) of their assigned group is minimized. 

### Exercise 5: Assigning Points to Clusters
Calculate the Euclidean distance between a data point and two cluster centroids, then assign the point to the closest cluster.

```python
import math

def euclidean_distance(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    dist = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return dist

# Given coordinates
data_point = (3, 4)
centroid_A = (0, 0)
centroid_B = (8, 8)

dist_A = euclidean_distance(data_point, centroid_A)
dist_B = euclidean_distance(data_point, centroid_B)

print(f"Distance to A: {dist_A:.2f}")
print(f"Distance to B: {dist_B:.2f}")

if dist_A < dist_B:
    assigned_cluster = "A"
else:
    assigned_cluster = "B"


print(f"The data point is assigned to Cluster: {assigned_cluster}")
```

## Section 7: Neural Networks and Deep Learning
Neural networks act as complex, non-linear classifiers. They consist of an **Input Layer**, one or more **Hidden Layers**, and an **Output Layer**.

Each node combines values from its inputs using a weight function. To introduce non-linearity (which allows the network to learn complex patterns), we pass the weighted sum through an activation function, such as the **Sigmoid function**.

### Exercise 7: Forward Pass of a Single Neuron
Calculate the weighted sum of inputs and pass it through the sigmoid activation function.

```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

# A single neuron receives 3 inputs from the previous layer
inputs = [0.5, 0.8, -0.2]
weights = [0.4, 0.6, -0.5] # The weights learned during training (backpropagation)
bias = 0.1

# calc of weighred sum
weighted_sum = 0
weighted_sum = 0
for i in range(len(inputs)):
    weighted_sum += inputs[i] * weights[i]

weighted_sum += bias

neuron_output = sigmoid(weighted_sum)

print(f"Weighted Sum: {weighted_sum}")
print(f"Neuron Activation Output: {neuron_output:.4f}")
```
